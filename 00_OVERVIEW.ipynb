```python
reg = requests.get("https://search.joins.com/TotalNews?page=159&Keyword=%EC%82%AC%EC%84%A4%20%EB%B0%95%EA%B7%BC%ED%98%9C&PeriodType=DirectInput&StartSearchDate=12%2F20%2F2012%2000%3A00%3A00&EndSearchDate=08%2F02%2F2018%2000%3A00%3A00&SortType=New&SourceGroupType=Joongang&ServiceCode=20&SearchCategoryType=TotalNews&MatchKeyword=%EC%82%AC%EC%84%A4")
response = TextResponse(reg.url, body=reg.text, encoding="utf-8")
response.xpath('//*[@id="content"]/div[3]/div[2]/ul/li/div/strong/a/text()')
```

```python
driver = webdriver.Chrome()
```

```python
url = "https://search.joins.com/TotalNews?page=159&Keyword=%EC%82%AC%EC%84%A4%20%EB%B0%95%EA%B7%BC%ED%98%9C&PeriodType=DirectInput&StartSearchDate=12%2F20%2F2012%2000%3A00%3A00&EndSearchDate=08%2F02%2F2018%2000%3A00%3A00&SortType=New&SourceGroupType=Joongang&ServiceCode=20&SearchCategoryType=TotalNews&MatchKeyword=%EC%82%AC%EC%84%A4"
driver.get(url)

items = driver.find_elements_by_css_selector("div.section_news > div.bd > ul > li")
```

```python
dict_list = []

for item in items:
    dict_list.append({
        "date": item.find_element_by_css_selector("div > span.byline > em:nth-child(2)").text,
        "headline": item.find_element_by_css_selector("div > strong.headline").text,
        "lead": item.find_element_by_css_selector("div > span.lead").text,
    })
    
df = pd.DataFrame(dict_list)    
df
```

```python
result_df = pd.DataFrame(columns=['date', 'headline', 'lead'])
result_df
```

```python
for i in range(1, 160):
    url = "https://search.joins.com/TotalNews?page=" + str(i) + "&Keyword=%EC%82%AC%EC%84%A4%20%EB%B0%95%EA%B7%BC%ED%98%9C&PeriodType=DirectInput&StartSearchDate=12%2F20%2F2012%2000%3A00%3A00&EndSearchDate=03%2F10%2F2017%2000%3A00%3A00&SortType=New&SourceGroupType=Joongang&ServiceCode=20&SearchCategoryType=TotalNews&MatchKeyword=%EC%82%AC%EC%84%A4"
    driver.get(url)

    items = driver.find_elements_by_css_selector("div.section_news > div.bd > ul > li")

    dict_list = []

    for item in items:
        dict_list.append({
            "date": item.find_element_by_css_selector("div > span.byline > em:nth-child(2)").text,
            "headline": item.find_element_by_css_selector("div > strong.headline").text,
            "lead": item.find_element_by_css_selector("div > span.lead").text,
        })

    df = pd.DataFrame(dict_list)    
    result_df = pd.concat([result_df, df])
```
